```python
# load libraries
import numpy as np
from matplotlib import pyplot as plt
from scipy.ndimage import measurements
```

# Perkolation
Im ersten Schritt erstellen wir ein Gitter mit $L\times L$ Gitterplätzen. Jeder Gitterplatz ist besetzt mit einer Zufallszahl zwischen 0 und 1.

```python
# linear size of lattice
L = 100

r = np.random.rand(L,L)
```

Nun überprüfen wir, welche Gitterplätze mit einer Zufallszahl kleiner als $p$ besetzt ist.

```python
# occupation probability
p = 0.55

lattice = r < p
```

Das resultierende Gitter mit binären (0 und 1) Zuständen plotten wir.

```python
plt.figure()
plt.imshow(lattice, 'binary', interpolation='nearest')
plt.xticks([])
plt.yticks([])
plt.show()
```

Jetzt können wir noch die Clustergrössen bestimmen...

```python
lw, num = measurements.label(lattice)

area = measurements.sum(lattice, lw, index=np.arange(lw.max() + 1))
areaImg = area[lw]
```

...und auch diese Resultate plotten.

```python
plt.figure()
plt.imshow(areaImg, 'jet', interpolation='nearest')
plt.xticks([])
plt.yticks([])
plt.show()
```
