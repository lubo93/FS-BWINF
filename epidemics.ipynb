```python
# load libraries
import networkx as nx
import EoN
import matplotlib.pyplot as plt
```

# Epidemics
Wir nutzen ein "SIR Modell", um die Ausbreitung einer Epidemie auf einem Gitter zu simulieren.

```python
# lattice with linear system size L
L = 100
G = nx.grid_2d_graph(L, L)
```

Wir benötigen eine gewisse Anzahl an anfänglichen Infektionen.

```python
initial_infections = [(u,v) for (u,v) in G if L//2-5<u<L//2+5 and L//2-5<v<L//2+5]
```

```python
pos = {node:node for node in G}

sim_kwargs = {'pos': pos}
```

Die Infektionsrate $\beta$ und Heilungsrate $\gamma$ definieren wir hier.

```python
# infection and recovery rates
beta = 2.0
gamma = 1.0
```

Jetzt müssen wir nur noch die Simulation starten und die Ergebnisse plotten.

```python
sim = EoN.fast_SIR(G, beta, gamma, initial_infecteds = initial_infections,
               tmax = 40, return_full_data=True, sim_kwargs = sim_kwargs)
```

```python
ani=sim.animate(ts_plots=['I', 'SIR'], node_size = 4, frame_times=[10])
```
